```python
import requests
import os
import zipfile

url = 'http://mlg.ucd.ie/files/datasets/bbc-fulltext.zip'

# Directory to save the downloaded dataset
download_dir = 'bbc_news_dataset'

# Create directory if it doesn't exist
if not os.path.exists(download_dir):
    os.makedirs(download_dir)

# Download the dataset
print(f"Downloading BBC News Articles dataset from {url}...")
response = requests.get(url)
file_path = os.path.join(download_dir, 'bbc-fulltext.zip')
with open(file_path, 'wb') as f:
    f.write(response.content)

# Extract the dataset
print("Extracting dataset...")
with zipfile.ZipFile(file_path, 'r') as zip_ref:
    zip_ref.extractall(download_dir)

print("Dataset downloaded and extracted.")
dataset_dir = 'bbc_news_dataset'

# Function to load the dataset
def load_bbc_dataset(directory):
    documents = []
    # print(os.listdir(directory))
    categories = os.listdir(os.path.join(directory,os.listdir(directory)[1]))

    for category in categories:
        category_dir = os.path.join(os.path.join(directory,os.listdir(directory)[1]), category)
        if not os.path.isdir(category_dir):
            continue
        print(category_dir)
        files = os.listdir(category_dir)
        for file in files[:10]:
            if file.endswith('.txt'):
                file_path = os.path.join(category_dir, file)
                with open(file_path, 'r', encoding='latin1') as f:
                    documents.append(f.read())

    return documents
# Load the dataset
documents = load_bbc_dataset(dataset_dir)
print(f"loaded {len(documents)} documents")
```

**BUILDING CORPUS**

```python
vocabulary = set(word for doc in tokenized_data for word in doc)
vocabulary = list(vocabulary)

# Create word_to_id and id_to_word dictionaries
word_to_id = {word: idx for idx, word in enumerate(vocabulary)}
id_to_word = {idx: word for word, idx in word_to_id.items()}

# Function to build corpus from tokenized data
def build_corpus(tokenized_documents, word_to_id):
    corpus = []
    for tokens in tokenized_documents:
        doc_ids = [word_to_id[word] for word in tokens if word in word_to_id]
        corpus.append(doc_ids)
    return corpus

# Build the corpus
corpus = build_corpus(tokenized_data, word_to_id)

# Print the corpus
print("Corpus:")
for i, doc in enumerate(corpus):
    print(f"Document {i+1}:", doc)
```

**LDA AND OUTPUT**

```python
import numpy as np
import random
import matplotlib.pyplot as plt
from gensim.models.coherencemodel import CoherenceModel
from gensim.corpora.dictionary import Dictionary


def LDA_Collapsed_Gibbs(corpus, num_topics, alpha, beta, num_iter=500):
    # Initialization
    num_docs = len(corpus)
    vocab_size = len(word_to_id)
    Z = [np.random.randint(low=0, high=num_topics, size=len(doc)) for doc in corpus]
    ndk = np.zeros((num_docs, num_topics))
    nkw = np.zeros((num_topics, vocab_size))
    nk = np.zeros(num_topics)

    # Count matrices
    for d in range(num_docs):
        for i, word in enumerate(corpus[d]):
            topic = Z[d][i]
            ndk[d, topic] += 1
            nkw[topic, word] += 1
            nk[topic] += 1

    # Gibbs Sampling
    for _ in range(num_iter):
        for d in range(num_docs):
            for i, word in enumerate(corpus[d]):
                topic = Z[d][i]
                ndk[d, topic] -= 1
                nkw[topic, word] -= 1
                nk[topic] -= 1

                # Calculate topic probabilities
                p_z = (ndk[d, :] + alpha) * (nkw[:, word] + beta) / (nk + beta * vocab_size)
                p_z /= np.sum(p_z)

                # Sample new topic
                new_topic = np.random.choice(num_topics, p=p_z)
                Z[d][i] = new_topic
                ndk[d, new_topic] += 1
                nkw[new_topic, word] += 1
                nk[new_topic] += 1

    # Calculate phi and theta
    theta = (ndk + alpha) / (np.sum(ndk, axis=1, keepdims=True) + num_topics * alpha)
    phi = (nkw + beta) / (nk[:, np.newaxis] + vocab_size * beta)

    return Z, ndk, nkw, nk, phi, theta



def calculate_coherence(phi, tokenized_data, word_to_id, id_to_word, measure='c_v'):
    topics = []
    for topic in phi:
        top_words = [id_to_word[idx] for idx in np.argsort(topic)[::-1] if idx in id_to_word]
        topics.append(top_words)

    texts = [[word for word in doc if word in word_to_id] for doc in tokenized_data]

    # Create a Gensim Dictionary from the texts
    gensim_dict = Dictionary(texts)

    # Filter out words that are not in the dictionary
    texts = [[word for word in doc if word in gensim_dict.token2id] for doc in texts]

    # Create the CoherenceModel
    coherence_model = CoherenceModel(topics=topics, texts=texts, dictionary=gensim_dict, coherence=measure)
    coherence_score = coherence_model.get_coherence()
    return coherence_score

# Assuming `phi` is the topic-term distribution matrix obtained from LDA
# `tokenized_data` is the tokenized documents (list of lists of tokens)
# `word_to_id` and `id_to_word` dictionaries are created from the vocabulary

# Initialize best_coherence and best_params
best_coherence = -np.inf
best_params = None

# Define parameter grids for tuning
num_topics_list = [5, 10, 15]  # example values
alphas = [0.01, 0.1, 1]
betas = [0.01, 0.1, 1]
num_iters = [200]

# Loop through parameter combinations
for num_topics in num_topics_list:
    for alpha in alphas:
        for beta in betas:
            for num_iter in num_iters:
                # Run LDA with the current set of parameters
                Z, ndk, nkw, nk, phi, theta = LDA_Collapsed_Gibbs(corpus, num_topics, alpha, beta, num_iter)

                # Calculate the coherence score for the current model
                coherence = calculate_coherence(phi, tokenized_data, word_to_id, id_to_word)
                print(f"Coherence: {coherence:.4f} | Topics: {num_topics}, Alpha: {alpha}, Beta: {beta}, Iterations: {num_iter}")

                # Update best_coherence and best_params if current coherence is better
                if coherence > best_coherence:
                    best_coherence = coherence
                    best_params = (num_topics, alpha, beta, num_iter)

# Print the best coherence score and parameters
print(f"Best Coherence: {best_coherence:.4f}")
print(f"Best Parameters: Topics: {best_params[0]}, Alpha: {best_params[1]}, Beta: {best_params[2]}, Iterations: {best_params[3]}")
```

**PLOTTING TOPICS DISTRIBUTION**

```python
# Function to get topic-word distribution as lists
def get_topic_word_distribution(phi, id_to_word, top_n=10):
    topics_words = []
    for topic_idx, topic_dist in enumerate(phi):
        word_distribution = [(id_to_word[i], prob) for i, prob in enumerate(topic_dist)]
        word_distribution.sort(key=lambda x: x[1], reverse=True)
        top_words = [word for word, prob in word_distribution[:top_n]]
        topics_words.append(top_words)
    return topics_words

# Example usage assuming phi and id_to_word are defined elsewhere
topics_words = get_topic_word_distribution(phi, id_to_word)

# Print the topics and top words in a list format
for topic_idx, top_words in enumerate(topics_words):
    print(f"Topic {topic_idx}: {top_words}")
```
